Graph adjacency list
- dynamically add or remove edges or vertices
- suitable for sparse graphs

```python
class graph:
    def __init__(self):
        self.adjList = {}
        
    def add_vertex(self,vertex):
        if vertex not in self.adjList:
            self.adjList[vertex] = []


    def add_edge(self,src,dest):
        self.add_vertex(src)
        self.add_vertex(dest)
        self.adjList[src].append(dest)
        self.adjList[dest].append(src) # if bidirectional, use this line

    def __str__(self):
        res = ""
        for vertex in self.adjList:
            res = res + str(vertex) + " -> " + str(self.adjList[vertex]) + "\n"
        return res
```

```python
g = graph()
g.add_edge(1,2)
g.add_edge(2,3)
g.add_edge(1,4)
g.add_edge(4,3)
g.add_edge(2,4)
g.add_edge(4,5)
g.add_edge(3,5)
```
