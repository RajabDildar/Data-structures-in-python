```python
class Node:
    def __init__(self,value,left=None,right=None):
        self.value = value
        self.left = left
        self.right = right
```

```python
# Making sample tree
root = Node(1)
root.left = Node(3)
root.right = Node(5)
root.left.left = Node(2)
root.left.right = Node(4)
root.right.right = Node(8)
```

- Preorder: root -> left -> right
- Inorder: left -> root -> right
- Post order: left -> right -> root

```python
def pre_order(root:Node):
    if root != None:
        print(root.value)
        pre_order(root.left)
        pre_order(root.right)

def in_order(root:Node):
    if root != None:
        in_order(root.left)
        print(root.value)
        in_order(root.right)

def post_order(root:Node):
    if root != None:
        post_order(root.left)
        post_order(root.right)
        print(root.value)
```

```python
print("Pre order")
pre_order(root)
print("In order")
in_order(root)
print("Post order")
post_order(root)
```
