Matrix representation
- suitable for dense graphs (where no. of edges are high)
- simple to code
- for less no. of edges, there would be wastage of memory.

```python
# queue is required for BFS
from queue import deque

class graph:
    def __init__(self,vertex):
        self.matrix = [[0]*vertex for _ in range(vertex)]
        self.size = vertex

    def add_edge(self,src, dest,weight=1):
        if 0 <= src < self.size and 0 <= dest < self.size:
            self.matrix[src][dest] = weight
            self.matrix[dest][src] = weight # if bidirectional, add this line also. if no directions, skip this line.
        else:
            print("Invalid edge.")

    def __str__(self):
        res = ""
        for row in self.matrix:
           res = res + ' '.join(map(str,row)) + "\n"
        return res
    
    # DFS (depth first search)
    def DFS(self,src):
        visited = [False]*self.size
        stack = [src]

        while (stack):
            vertex = stack.pop()

            if visited[vertex] == False:
                print(vertex,end=" ")
                visited[vertex] = True

            for i in range(self.size):
                if self.matrix[vertex][i] == 1 and visited[i] == False:
                    stack.append(i)

    # BFS (breadth first search)
    def BFS(self,src):
        visited = [False] * self.size
        q = deque([src])
        visited[src] = True
        
        while (q):
            v = q.popleft()
            print(v,end=" ")

            for i in range(self.size):
                if (self.matrix[v][i] == 1 and visited[i] == False):
                    visited[i] = True
                    q.append(i)
```

```python
g = graph(5)
g.add_edge(0,1)
g.add_edge(0,2)
g.add_edge(1,3)
g.add_edge(2,4)
g.add_edge(3,4)
g.add_edge(2,3)
```
