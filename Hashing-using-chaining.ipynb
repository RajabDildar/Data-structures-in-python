Hashing via Chaining

```python
class Node:
    def __init__(self,key,value,next=None):
        self.key = key 
        self.value = value 
        self.next = next
```

```python
class LinkedList:
    def __init__(self):
        self.head = None
        # no. of nodes in LinkedList
        self.n = 0

    def __len__(self):
        return self.n
    
    def __str__(self):
        return self.traverse()

    def traverse(self):
        curr_node = self.head
        result = ""
        while curr_node != None:
            result = result + str(curr_node.key) + ": " + str(curr_node.value) + ","
            curr_node = curr_node.next

        return result[:-1]

    def append(self,key,value):
        new_node = Node(key,value)

        if self.head == None:
            self.head = new_node
            self.n += 1
            return
        
        curr_node = self.head
        while curr_node.next != None:
            curr_node = curr_node.next
            
        curr_node.next = new_node
        self.n += 1

    def delete_head(self):
        if (self.head == None):
            return "Empty linkedlist"
        self.head = self.head.next
        self.n -= 1

    def remove(self,key):
        curr_node = self.head

        if curr_node == None:
            return "Empty LL"

        if curr_node.key == key:
            self.delete_head()
            return

        while curr_node.next != None:
            if(curr_node.next.key == key):
                break
            curr_node = curr_node.next
        else:
            return "value not found"
        
        curr_node.next = curr_node.next.next
        self.n -= 1

    def search(self, key):
        index = 0
        curr_node = self.head

        #traverse
        while curr_node != None:
            if(curr_node.key == key):
                break
            curr_node = curr_node.next
            index += 1
        else:
            return -1
        
        return index

    def get_node(self,index):
        curr_node = self.head
        for i in range(index):
            if(curr_node != None):
                curr_node = curr_node.next
            else:
                return "Index Not found"
        
        if (curr_node != None):
            return curr_node
        else:
            return "Index Not found"
```

```python
class Dictionary:
    def __init__(self,capacity=4):
        self.capacity = capacity
        self.size = 0
        # create array of Linked List
        self.buckets = self.__make_array(capacity)

    def __make_array(self,capacity):
        buckets = []
        for i in range(capacity):
            buckets.append(LinkedList())        
        return buckets
    
    @property
    def laod_factor(self):
        return self.size/self.capacity
    
    def __getitem__(self,key):
        return self.get(key)

    def __setitem__(self,key,value):
        return self.put(key,value)
    
    def __delitem__(self,key):
        bucket_index = self.hash_function(key)
        bucket = self.buckets[bucket_index]
        bucket.remove(key)
        self.size-=1

    def __str__(self):
        res = ""
        for i in self.buckets:
            res += i.traverse()  + ","
        return "{ " + res + " }"
    
    def get(self,key):
        bucket_index = self.hash_function(key)
        bucket = self.buckets[bucket_index]
        res = bucket.search(key)

        if res == -1:
            return "Not found"
        else:
            node = bucket.get_node(res)
            return node.value
    
    def put(self,key,value):
        bucket_index = self.hash_function(key)
        bucket = self.buckets[bucket_index]
        node_index = self.get_node_index(bucket_index,key)
        if node_index == -1:
            # not found, append
            bucket.append(key,value)
            self.size+=1
            if (self.laod_factor >= 2):
                self.rehash()
        else:
            # found, edit
            node = bucket.get_node(node_index)
            node.value = value

    def rehash(self):
        self.capacity *= 2
        self.size = 0
        prev_buckets = self.buckets
        self.buckets = self.__make_array(self.capacity)

        for prev_bucket in prev_buckets:
            for i in range(len(prev_bucket)):
                node = prev_bucket.get_node(i)
                node_key = node.key
                node_value = node.value
                self.put(node_key,node_value)

    def get_node_index(self,bucket_index,key):
        bucket = self.buckets[bucket_index]
        node_index = bucket.search(key)
        return node_index
    
    def hash_function(self,key):
        return abs(hash(key)) % self.capacity
```
